# Парсинг сайта [MyAnimeList](https://myanimelist.net/)

Сначала напишем функцию, чтобы достать свойства, т.к. они находятся на одном уровне и имею одни и те же теги с классами.

Под свойствами тут понимаются студии, источник, темы и ориентацию на пол и возраст зрителей.

```python
def get_propertys(propertys):
  title_propertys = {'Studios': None, 'Source': None, 'Themes': None, 'Demographic': None}
  for property in propertys:
    if (property.span.text.strip() == 'Studios') or (property.span.text.strip() == 'Studio'):
      anime_studios = property.find_all('span', {'class': 'item'})
      title_propertys['Studios'] =  ", ".join(element.text for element in anime_studios)

    if property.span.text.strip() == 'Source':
      title_propertys['Source'] = property.span.next_element.next_element.text.strip()

    if (property.span.text.strip() == 'Themes') or (property.span.text.strip() == 'Theme'):
      anime_themes =  property.find_all('span', {'class': 'item'})
      title_propertys['Themes'] = ", ".join(element.text for element in anime_themes)

    if property.span.text.strip() == 'Demographic':
      title_propertys['Demographic'] = property.span.next_element.next_element.text.strip()

  return title_propertys
```

Теперь напишем функцию сбора информации из странички.

```python
def get_page(id_genre ,genre, p):

  url = f'https://myanimelist.net/anime/genre/{id_genre}/{genre}?page={p}'

  response = requests.get(url, headers={'User-Agent': UserAgent().chrome})

  tree = BeautifulSoup(response.content, 'html.parser')

  anime = tree.find_all('div', {'class': 'js-anime-category-producer seasonal-anime js-seasonal-anime js-anime-type-all js-anime-type-1'})
  anime.extend(tree.find_all('div', {'class': 'js-anime-category-producer seasonal-anime js-seasonal-anime js-anime-type-all js-anime-type-2'}))
  anime.extend(tree.find_all('div', {'class': 'js-anime-category-producer seasonal-anime js-seasonal-anime js-anime-type-all js-anime-type-3'}))
  anime.extend(tree.find_all('div', {'class': 'js-anime-category-producer seasonal-anime js-seasonal-anime js-anime-type-all js-anime-type-4'}))
  anime.extend(tree.find_all('div', {'class': 'js-anime-category-producer seasonal-anime js-seasonal-anime js-anime-type-all js-anime-type-5'}))
  anime.extend(tree.find_all('div', {'class': 'js-anime-category-producer seasonal-anime js-seasonal-anime js-anime-type-all js-anime-type-6'}))
  anime.extend(tree.find_all('div', {'class': 'js-anime-category-producer seasonal-anime js-seasonal-anime js-anime-type-all js-anime-type-7'}))

  info = []

  for title in anime:
      # Studios, Source, themes, demographic
      propertys = title.find_all('div', {'class': 'property'})
      title_propertys = get_propertys(propertys)

      title_Name = title.find('a', {'class': 'link-title'}).text.strip()
      score = title.find('span', {'class': 'js-score'}).text.strip()
      url = title.find('a', {'class': 'link-title'}).get('href')
      genres = title.find('div', {'class': 'genres-inner js-genre-inner'}).text.strip().replace('\n\n', ', ')
      studios = title_propertys['Studios']
      source = title_propertys['Source']
      members = title.find('span', {'class': 'js-members'}).text.strip()
      synopsis = title.find('p', {'class': 'preline'}).text.strip().replace('\r\n\r\n', ' ')
      year = title.find('div', {'class': 'info'}).span.text.strip()[-4:]
      themes = title_propertys['Themes']
      demographic = title_propertys['Demographic']
      ep = title.find('div', {'class': 'info'}).span.next_element.next_element.next_element.next_element.text.strip()[:2]

      data = {
          'title_Name': title_Name,
          'score': score,
          'url': url,
          'genres': genres,
          'studios': studios,
          'source': source,
          'members': members,
          'synopsis': synopsis,
          'year': year,
          'themes': themes,
          'demographic': demographic,
          'ep': ep
          }

      info.append(data)

  return info
```

Соберем данные с сайта.

```python
SLEEP = 2 # время ожидания перед след. запросом
DEPTH = 15 # кол-во страниц каждого жанра
GENRES = [["Action", "Adventure", "Comedy", "Drama", "Fantasy", "Romance", "Sci-Fi"],
          [1, 2, 4, 8, 10, 22, 24]]
```

```python
info = []

for count_genres in range(len(GENRES[0])):
  for number_page in range(1, DEPTH + 1):
    time.sleep(SLEEP)

    try:
      info.extend(get_page(GENRES[1][count_genres], GENRES[0][count_genres], number_page))

    except:
      pass

    #сохранение промежуточных
    if number_page % 5 == 0:
      df = pd.DataFrame(info)
      df.to_csv(f'anime_{GENRES[0][count_genres]}_{number_page}.csv', index=False)
```

Сохраним полностью собранные результаты

```python
df = pd.DataFrame(info)
df['score'] = df['score'].astype(float)
df['members'] = df['members'].astype(int)

df['year'] = df['year'].apply(lambda x: re.sub("[^0-9]", "", x))
df.loc[(df['year'] == ""), 'year'] = None
#df['year'] = df['year'].astype(int)
df['ep'] = df['ep'].apply(lambda x: re.sub("[^0-9]", "", x))
df.loc[(df['ep'] == ""), 'ep'] = None
#df['ep'] = df['ep'].astype(int)

df.to_csv(f'anime_result.csv', index=False)
```
