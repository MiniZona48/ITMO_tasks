```python
import pandas as pd
import numpy as np
from sklearn.neighbors import KNeighborsClassifier

df0 = pd.read_csv("Data/import.csv", index_col=0)

knn0 = KNeighborsClassifier(n_neighbors=3, p=2)

knn0.fit(df0[["X", "Y"]], df0["Class"])

#  ,      0
print(knn0.kneighbors(np.array([[58, 63]]), return_distance=True))

print(knn0.predict(np.array([[58, 63]])))

knn0 = KNeighborsClassifier(n_neighbors=3, p=1)

knn0.fit(df0[["X", "Y"]], df0["Class"])

#  ,      0
print(knn0.kneighbors(np.array([[58, 63]]), return_distance=True))

print(knn0.predict(np.array([[58, 63]])))
```
