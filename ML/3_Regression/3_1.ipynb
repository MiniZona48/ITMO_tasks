```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

#загружаю дата сет
data = pd.read_csv('Datasets/task1.csv', index_col=False, header=0)

X = data.X.values
Y = data.Y.values
length = len(data)

X = X.reshape(length, 1)
Y = Y.reshape(length, 1)

print('Выборочное среднее:', data.mean(axis=0))

# Создаем модель
model = LinearRegression()

# Совмещение модели
model.fit(X, Y)

# вычисляем R2
R2 = model.score(X,Y)
print('coefficient of determination:', R2)

# получаем тета
b0 = model.intercept_ # тета 0
b1 = model.coef_ # тета 1

print('intercept:', b0)
print('coef:', b1)

plt.scatter(X, Y,  color='black')
plt.plot(X, model.predict(X), color='blue', linewidth=3)
plt.xticks(())
plt.yticks(())
plt.show()
```
